## 2.파이썬 패키지 : Pillow

````python
from PIL import Image
 
# 이미지 읽기
file_name = '../cv_images/banana.jpg'
image = Image.open(file_name)
 
# 이미지 보여주기
image.show()

# 이미지 저장 : Image.save(File Path, format=None)
image.save("Test.bmp",'BMP') 

# 이미지 데이터 배열로 저장하기
import numpy as np
a = np.array(image)
a[:2]
````
