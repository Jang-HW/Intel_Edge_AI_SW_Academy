### 2.3.2 결측치 채우기

- `fillna()` Imputation - 대체값 넣기  , 결측치에  **mean** 또는 **median**을 채워준다.

````python
# .mean() 사용하여 revenue_millions 컬럼 평균구하기
revenue_mean = df["revenue_millions"].mean()          #... 코드 작성

revenue_mean
````

- Null 값에 `fillna()` 메서드로 평균값을 채우기
- `inplace=True` 인자를 넣어서 원래  `df`까지 적용 시킨다

````python
# revenue Null 값에 fillna() 메서드로 평균값을 채우고 revenue에 반영하기
df["revenue_millions"].fillna(revenue_mean, inplace=True)
````

````python
df.isnull().sum()
````

````python
# .mean() 사용하여 metascore 컬럼 평균구하기
metascore_mean = df["metascore"].mean()          #... 코드 작성

metascore_mean
````

````python
# metascore Null 값에 fillna() 메서드로 평균값을 채우고 metascore 반영하기
df["metascore"] = df["metascore"].fillna(metascore_mean)               #... 코드 작성
````

`keep=False`,은 모든 중복 데이터를 삭제한다. 복제 후 두 데이터가 동일하면 keep=False 인자는 모든 데이터를 삭제한다.

- Keep 의 인자의 사용에 따라 ,
    * `first`: 디폴트로 첫번째 데이터만 빼고 drop 한다. default
    * `last`: 마지막 데이터만 빼고 drop 한다.
    * `False`: 처음인지 끝값인지 고려 안하고 중복이면 무조건 Ture, 모든 데이터를 삭제한다.

````python
temp_df = pd.concat([df, df], ignore_index=True)

temp_df.drop_duplicates(inplace=True, keep=False)

temp_df.shape
````

## 4. 유용한 기능 Apply( ) 기능 사용하기

- DataFrame과 Series는 리스트로 사용하기 때문에, Dataset이 커지면 연산 속도가 느려진다.

-  `apply()` 함수를 사용하면 복수의 Function을 빠르게 적용할 수 있다.
- pandas는 Vectorization에 최적화 됐기 때문에`apply()`를 사용하면 수동으로 사용하는 반복문 보다 훨씬 빠른 속도로 연산할 수 있다.

````python
def rating_function(x):
    if x >= 8.0:
        return "good"
    else:
        return "bad"
````

````python
df['rating_category'] = df['rating'].apply(rating_function)

df.head(5)
````

- `.apply()`와 `lambda`를 조합해서 사용하면 for문을 함수처럼 적용 가능하다.
- 파이썬에서 "lambda" 는 런타임에 생성해서 사용할 수 있는 익명 함수이다.
- lambda는 쓰고 버리는 일시적인 함수로 생성된 곳에서만 적용한다.

````python
df['rating_category'] = df['rating'].apply(lambda x: 'good' if x >= 8.0 else 'bad')

df.head(5)
````
