**4.텍스트 음성 변환**

```python
text = "안녕. 파이썬 수업입니다."

tts = gtts.gTTS(text=text, lang='ko')
tts.save("tts1.mp3")
```

**4.2 텍스트 파일 읽어 음성 변환**

```python
file_name = 'TTS텍스트.txt'
with open(file_name, 'rt', encoding='UTF8') as f :
    read_file = f.read()

tts = gtts.gTTS(text=read_file, lang='ko')
tts.save("myText.mp3")

wn = IPython.display.Audio('myText.mp3', autoplay=True)
IPython.display.display(wn)
```
